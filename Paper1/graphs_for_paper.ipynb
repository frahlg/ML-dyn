```python
import matplotlib.pyplot as plt

date1='2014-04-05'
date2='2014-04-06'

dates = date1 + ':' + date2

results_folder = 'results/paper_180209/'

intervals = [[0,4],[4,8],[8,14],[14,15],[15,19],[19,23],[23,29],[29,30]]

runs = ['linear','tpot','autosk']

for interval_ in intervals:
    for run_ in runs:

        fig, ax = plt.subplots()
        fig.set_size_inches(8, 5)

        if interval_[0] < 15:
            plt.plot(df_out['y_13'][date1:date2],label='Measured value engine 1_3',marker='x')
        if interval_[0] >= 15:
            plt.plot(df_out['y_24'][date1:date2],label='Measured value engine 2_4',marker='x')
        
        for i in range(interval_[0],interval_[1]):
            plt.plot(df_out[run_ + '_test_no_'+str(i)][date1:date2],label=run_ + ' test '+str(i))
        # ,marker='.'
        if run_ == 'linear':
            ax.set(xlabel='Time', ylabel='FO Flow',title='Linear regression')
        if run_ == 'tpot':
            ax.set(xlabel='Time', ylabel='FO Flow',title='TPOT regression 10 Gen')
        if run_ == 'autosk':
            ax.set(xlabel='Time', ylabel='FO Flow',title='Auto-sklearn')

        ax.grid()
        plt.legend(loc='upper right')
        plt.xlim(pd.Timestamp(date1),pd.Timestamp(date2))
        plt.show()
        fig.savefig(results_folder + run_ + str(interval_) + date1 + "_test.png",dpi=1200)
#
```

```python

# Plot difference instead of absolute values.

df_out_diff = pd.DataFrame()

df_out_diff['y_13'] = df_out['y_13']
df_out_diff['y_24'] = df_out['y_24']

runs = ['linear','tpot','autosk']

for run_ in runs:
    for f in range(0,15):
        df_out_diff[run_ + '_test_no_'+str(f)] = (df_out[run_ + '_test_no_'+str(f)] - df_out['y_13'])/max(df_out['y_13'])
    for f in range(15,30):
        df_out_diff[run_ + '_test_no_'+str(f)] = (df_out[run_ + '_test_no_'+str(f)] - df_out['y_24'])/max(df_out['y_24'])

df_out_diff
```

```python

# Plot difference instead of absolute values.

import matplotlib.pyplot as plt

date1='2014-04-05'
date2='2014-04-06'

results_folder = 'results/paper_180209/'

intervals = [[0,4],[4,8],[8,14],[14,15],[15,19],[19,23],[23,29],[29,30]]

runs = ['linear','tpot','autosk']

for interval_ in intervals:
    for run_ in runs:

        fig, ax = plt.subplots()
        fig.set_size_inches(8, 5)

        #plt.plot(df_out['y_13'][date1],label='Measured value',marker='x')
        
        for i in range(interval_[0],interval_[1]):
            plt.plot(df_out_diff[run_ + '_test_no_'+str(i)][date1:date2],label=run_ + ' test '+str(i))
        # ,marker='.'
        if run_ == 'linear':
            ax.set(xlabel='Time', ylabel='FO Flow difference',title='Linear regression')
        if run_ == 'tpot':
            ax.set(xlabel='Time', ylabel='FO Flow difference',title='TPOT regression 10 Gen')
        if run_ == 'autosk':
            ax.set(xlabel='Time', ylabel='FO Flow difference',title='Auto-sklearn')
          
        ax.grid()
        ax.set_ylim(-0.1,0.1)
        plt.legend(loc='upper right')
        plt.xlim(pd.Timestamp(date1),pd.Timestamp(date2))
        plt.show()
        fig.savefig(results_folder + run_ + str(interval_) + date1 + "_diff_test.png",dpi=1200)

        
# plot test 15 for all comparison


runs = ['linear','tpot','autosk']
intervals = [14,29]

for i in intervals:
    
    fig, ax = plt.subplots()
    fig.set_size_inches(8, 5)
    
    for run_ in runs:
        plt.plot(df_out_diff[run_ +'_test_no_'+str(i)][date1:date2],label=run_ + ' test '+str(i))

    ax.set(xlabel='Time', ylabel='FO Flow difference',title='Comparison Linear, TPOT and Auto-sklearn')
    ax.grid()
    ax.set_ylim(-0.05,0.05)
    plt.legend(loc='upper right')
    plt.xlim(pd.Timestamp(date1),pd.Timestamp(date2))
    plt.show()
    fig.savefig(results_folder + 'all_' + str(i) + date1 + "_diff_test.png",dpi=1200)
        
```
